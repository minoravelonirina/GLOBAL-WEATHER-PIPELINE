```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from datetime import datetime

# Configuration des chemins
processing_date = datetime.now()
date_str = processing_date.strftime("%Y-%m-%d")
project_root = Path().resolve().parent
data_dir = project_root / "data"
output_dir = data_dir / "processed" / date_str
input_file = output_dir / "weather_metrics.csv"

# Chargement des donn√©es
df = pd.read_csv(input_file)

# Nettoyage basique
df["rain"] = pd.to_numeric(df["rain"], errors="coerce")
df = df.dropna(subset=["rain", "city"])

# Affichage des infos g√©n√©rales
print("R√©sum√© statistique g√©n√©ral :")
print(df.describe())

print("\nNombre de lignes par ville :")
print(df["city"].value_counts())

# Agr√©gation : pluie totale par ville
rain_totals = df.groupby("city")["rain"].sum().sort_values(ascending=False)

print("\nTotal de pluie par ville (mm) :")
print(rain_totals)

# Ville la plus pluvieuse
most_rainy_city = rain_totals.idxmax()
max_rain = rain_totals.max()

print(f"\nüåßÔ∏è Ville la plus pluvieuse : {most_rainy_city} avec {max_rain:.2f} mm de pluie cumul√©e.")

# Visualisation
plt.figure(figsize=(10, 6))
sns.barplot(x=rain_totals.index, y=rain_totals.values, palette="Blues_d", legend=False)
plt.title("Total de pluie par ville (mm)")
plt.ylabel("Pluie cumul√©e (mm)")
plt.xlabel("Ville")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
